### Format data to run SpatialDM per FOV  
Each fov is understood as a technical sample

Create the new 'donor' field by combining 'condition' and 'slide'

````python
S1.obs['donor'] = S1.obs[['condition', 'slide']].apply(lambda row: f'{row["condition"]}_{row["slide"]}', axis=1)
S2.obs['donor'] = S2.obs[['condition', 'slide']].apply(lambda row: f'{row["condition"]}_{row["slide"]}', axis=1)
S3.obs['donor'] = S3.obs[['condition', 'slide']].apply(lambda row: f'{row["condition"]}_{row["slide"]}', axis=1)
````

Create the new 'sample' field by combining 'condition', 'slide' and 'fov'

````python
S1.obs['sample'] = S1.obs[['condition', 'slide', 'fov']].apply(lambda row: f'{row["condition"]}_{row["slide"]}_{row["fov"]}', axis=1)
S2.obs['sample'] = S2.obs[['condition', 'slide', 'fov']].apply(lambda row: f'{row["condition"]}_{row["slide"]}_{row["fov"]}', axis=1)
S3.obs['sample'] = S3.obs[['condition', 'slide', 'fov']].apply(lambda row: f'{row["condition"]}_{row["slide"]}_{row["fov"]}', axis=1)
````

````python
slides = [S1,S2,S3]
````

````python
samples = []
for s in range(3):
    slide = slides[s]
    print(s)
    for f in range(len(slide.obs.fov.unique())):
        fov = str(f+1)
        sample = slide[np.isin(slide.obs['fov'],[fov])].copy()
        samples.append(sample)
        del sample
````

````python
for i in range(3):
    for adata in samples:
        #adata.obsm['spatial'] = adata.obsm['spatial'].values
        #adata.X = adata.X.toarray()
        sdm.weight_matrix(adata, l=100, cutoff=0.2, single_cell=True) # weight_matrix by rbf kernel
        sdm.extract_lr(adata, 'human', min_cell=3)      # find overlapping LRs from CellChatDB
        sdm.spatialdm_global(adata, 1000, specified_ind=None, method='z-score', nproc=8)     # global Moran selection
        sdm.sig_pairs(adata, method='z-score', fdr=True, threshold=0.1)     # select significant pairs
        sdm.spatialdm_local(adata, n_perm=1000, method='z-score', specified_ind=None, nproc=8)     # local spot selection
        sdm.sig_spots(adata, method='z-score', fdr=False, threshold=0.1)
        print(' done!')
````

#### Number of interaction per condition and per donor

````python
data = []

for sample in samples:
    # Extracting fields and calculating the number of interactions
    sample_name = sample.obs['sample'].iloc[0]  # Assuming 'sample' is consistent for each object
    condition = sample.obs['condition'].iloc[0]  # Assuming a single condition per object
    donor = sample.obs['donor'].iloc[0]  # Assuming a single condition per object
    num_interactions = sample.uns['global_res']['selected'].sum()
    
    # Append to the data list
    data.append({'sample': sample_name, 'condition': condition, 'donor': donor,'number_of_interactions': num_interactions})

# Create a DataFrame from the data list
df = pd.DataFrame(data)

# Set 'sample' as the index
df.set_index('sample', inplace=True)
df
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Assuming df is already created and contains the necessary data

# Ensure the 'donor' column is ordered according to group_order
df['donor'] = pd.Categorical(df['donor'], categories=group_order, ordered=True)

# Perform One-way ANOVA
grouped = [group['number_of_interactions'].values for _, group in df.groupby('donor')]
anova_stat, anova_p = f_oneway(*grouped)

# If ANOVA shows significance, perform Tukey HSD test for pairwise comparisons
if anova_p < 0.05:
    print(f"ANOVA Results: F-statistic = {anova_stat:.3f}, p-value = {anova_p:.3e}")
    print("ANOVA shows a significant difference across donors. Performing pairwise comparisons.")
    
    # Perform Tukey HSD pairwise comparisons
    tukey = pairwise_tukeyhsd(df['number_of_interactions'], df['donor'])
    tukey_summary = tukey.summary()
    print(tukey_summary)

    # Visualize the Tukey HSD test results on the plot
    tukey_results = tukey_summary.data[1:]  # Extracting results (ignoring the header)
    
    # Plotting
    plt.figure(figsize=(4, 4))
    sns.barplot(data=df, x='donor', y='number_of_interactions', order=group_order, palette=group_colors)
    plt.xlabel('Donor')
    plt.ylabel('Number of Interactions')
    plt.title('Number of Interactions Across Donors')

    # Annotate bars with p-values from Tukey test
    for i, donor in enumerate(group_order):
        for result in tukey_results:
            if donor in result[0] and donor in result[1]:  # Matching the donor in the comparison
                p_val = result[4]  # Extract p-value
                plt.text(i, df[df['donor'] == donor]['number_of_interactions'].mean() + 0.2,
                         f"p = {p_val:.3e}",
                         ha='center', va='bottom', color='black')

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('interaction_plot_with_pvalues.pdf', format='pdf')
    #plt.close()

else:
    print(f"ANOVA Results: F-statistic = {anova_stat:.3f}, p-value = {anova_p:.3e}")
    print("ANOVA does not show a significant difference across donors.")
    
# Optional: Kruskal-Wallis test for non-parametric data
from scipy.stats import kruskal
kruskal_stat, kruskal_p = kruskal(*grouped)
print(f"Kruskal-Wallis Results: H-statistic = {kruskal_stat:.3f}, p-value = {kruskal_p:.3e}")

````

````python
_range = np.arange(1, concat.uns['n_sub'])
test = 1-concat.uns['p_df'].loc[(concat.uns['p_val']<0.05) & (concat.uns['tf_df'].sum(1).isin(_range)),
                                     concat.uns['subset']]
````

````python
test.index
````

````python
# Define groups based on prefixes
group_order = [
    "T1D_Run5211_S1",  # T1D donor 1
    "T1D_Run5211_S2",  # T1D donor 2
    "T1D_Run5211_S3",  # T1D donor 3
    "ND_Run5211_S1",   # ND donor 1
    "ND_Run5211_S2",   # ND donor 2
    "ND_Run5211_S3"    # ND donor 3
]

# Group the columns by prefix
grouped_columns = {group: [col for col in test.columns if col.startswith(group)] for group in group_order}

# Calculate the average for each group
averaged_df = pd.DataFrame()

for group, cols in grouped_columns.items():
    # Calculate the mean for each group of columns
    averaged_df[group] = test[cols].mean(axis=1)

group_colors = {
    "T1D_Run5211_S1": "#a1c349",
    "T1D_Run5211_S2": "#87a330",
    "T1D_Run5211_S3": "#6a8532",
    "ND_Run5211_S1": "#fd5901",
    "ND_Run5211_S2": "#f78104",
    "ND_Run5211_S3": "#faab36",
}

# Create col_colors for the averaged DataFrame columns based on the group order
col_colors = [group_colors[group] for group in group_order]

# Create the clustermap with the colorbar on top
color_map = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
g = sns.clustermap(averaged_df, cmap=color_map, col_cluster=False, col_colors=col_colors)

# Add a legend for the group colors
from matplotlib.patches import Patch
legend_handles = [Patch(facecolor=color, label=group) for group, color in group_colors.items()]
plt.gcf().legend(handles=legend_handles, title="Groups", loc="upper center", bbox_to_anchor=(0.5, 1.05), ncol=2)

# Adjust the figure size to ensure all row names are visible
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0, fontsize=10)  # Rotate row labels for better visibility
plt.tight_layout()

# Save the clustermap as a PDF
g.savefig("clustermap_output.pdf", format='pdf')

# Show the plot
plt.show()
````
